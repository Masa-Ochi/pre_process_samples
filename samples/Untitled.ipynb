```python
df = pd.read_csv("./reserve.csv")
hotel = pd.read_csv("./hotel.csv")
```

```python
# 【集約】極値・代表値の算出
(
df
    .groupby('hotel_id')
    .agg({'total_price': ['max', 'min', 'mean', 'median', lambda x: np.percentile(x, q=20)]})
    .reset_index()
    .rename(columns = {
        'max': 'price_ max', 
        'min': 'price_ min', 
        'mean': 'price_ mean', 
        'median': 'price_ median', 
        '<lambda>': 'price_ 20 per'})
)
```

```python
# 【集約】分散値と標準偏差
(
df.groupby('hotel_id')
    .agg({'total_price': ['var', 'std']})
    .reset_index()
    .rename(columns={
         'var': 'price_var'
        ,'std': 'price_std'
    })
)
```

```python
# 【集約】最頻値の算出
(
df
    .loc[:, ['total_price']]
    .assign(
        round_price=lambda df: df.total_price.round(-3)
    )['round_price']
    .mode()
    .reset_index()
#     .plot.bar()
)
```

```python
# 【集約】順位の算出
(
df
    .groupby('hotel_id')
    .size()
    .reset_index()
    .rename(columns={
        0: 'rsv_cnt'
    })
    .assign(
        rsv_cnt_rank=lambda df: df.rsv_cnt.rank(ascending=False, method='min').astype(np.int64)
    )
    .loc[:, ['hotel_id', 'rsv_cnt_rank']]
    .sort_values('rsv_cnt_rank')
    .pipe(
#         lambda df: df[df.rsv_cnt_rank < 10]
        lambda df: df.query('rsv_cnt_rank < 10')
    )
)
```

```python
# 【集約】ランキングによるグルーピング
(
hotel
    .groupby(['big_area_name', 'small_area_name'], as_index = False) \
    .size()
    .reset_index()
    .rename(columns={
        0: 'hotel_cnt'
    })
    .assign(
        join_area_id=lambda df: np.where(df['hotel_cnt']-1 >= 20, 
                                         df['small_area_name'], 
                                         df['big_area_name'])
    )
    .loc[:, ['small_area_name', 'join_area_id']]
    .pipe(
        lambda df: pd.merge(df, hotel, on='small_area_name')
    )
    .loc[:, ['hotel_id', 'join_area_id']]
)
```

```python
# 【結合】n件前のデータ取得
(
df
    .groupby('customer_id')
    .apply(
        lambda df: df.sort_values(by='reserve_datetime', axis=0, inplace=False)
    )
    .assign(
        before_price=lambda df: pd.Series(df['total_price'].shift(periods=2))
    )
    .assign(
        before_price=lambda df: df['before_price'].fillna(0)
    )
)
```

```python
# 【結合】過去n件の合計値
(
df
    .assign(
        price_sum=lambda df: df
                                            .groupby('customer_id')
                                            .apply(
                                               lambda df: df.sort_values(by='reserve_datetime', ascending=True)
                                            )
                                            .loc[:, 'total_price']
                                            .rolling(center=False, window=3, min_periods=3).sum()
                                            .reset_index(drop=True)
    )
)
```

```python
# 【結合】過去n件の平均値
(
df
    .assign(
        price_avg=lambda df: df
                                            .groupby('customer_id')
                                            .apply(
                                                lambda df: df.sort_values(by='reserve_datetime', ascending=True)
                                            )
                                            .loc[:, 'total_price']
                                            .rolling(center=False, window=3, min_periods=3).mean()
                                            .reset_index(drop=True)
    )
    .assign(
        price_avg=lambda df: df
                                            .groupby('customer_id')['price_avg']
                                            .shift(periods=1)
    )
)
```

```python
# 【結合】過去n日間の合計値
n=-90
(
df
    .assign(
        reserve_datetime=lambda df: pd.to_datetime(df['reserve_datetime'], format='%Y-%m-%d %H:%M:%S')
    )
    .pipe(
        lambda df: pd.merge(
                                df[['reserve_id', 'customer_id', 'reserve_datetime']], 
                                df[['customer_id', 'reserve_datetime', 'total_price']]
                                    .rename(columns ={'reserve_datetime': 'reserve_datetime_before'}), 
                                on ='customer_id')
    )
    .pipe(
        lambda df: df[operator.and_( 
                                df['reserve_datetime'] > df['reserve_datetime_before'], 
                                df['reserve_datetime'] + offsets.Day(n) <= df['reserve_datetime_before']
                                )]
    )
    .groupby('reserve_id')['total_price'].sum()
    .reset_index()
    .rename(columns={
        'total_price': 'total_price_sum'
    })
    .pipe(
        lambda _df: pd.merge(df, _df, on='reserve_id', how='left')
    )
    .fillna(0)
)
```
